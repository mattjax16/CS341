## Part 2: Scipy Basics

In CS341, we will use scipy.linalg to find the distance between two vectors, scipy.signal to find the peaks in a time-series, and scipy.integrate to solve ordinary differential equations (i.e. simulate a model).

If you have a recent enough version of scipy, then it will have [scipy.integrate.solve_ivp](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.solve_ivp.html), which is short for "solve initial value problem". It takes as input the model, the initial value (the concentration of all the proteins at time = 0), the time steps over which you want to simulate the model, and parameter values. It returns the concentration of proteins at those time steps.

If you have an older version of scipy, then you will need to use [scipy.integrate.odeint](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.odeint.html#scipy.integrate.odeint), which takes the model in a different form (the order of arguments is different).

We are going to begin with the simplest model typically used. In this model, we have one protein $P$, whose concentration we will model over time. The process we will model is called degradation - the process by which the protein is destroyed. At any given moment in time, a given proportion of that protein will be destroyed. If we start with a lot of protein, then early in the simulation we will be losing lots of it (it will be degrading quickly). Once we have just a little protein, we lose less at any given moment, so the process is slower. We capture that with the ordinary differential equation

$\frac{dP(t)}{dt} = -k P(t)$

where k is the constant that determines the proportion of protein $P$ that is degraded. I read this as "the rate of change of P depends is proportional to the amount of P" and, if we assume that k>=0 (which is usually the case in these models), then we see that the rate of change of P is negative, and will be more negative (bigger) when P is bigger".

Below, use scipy.integrate.solve_vip to simulate the degradation model when $k=2$. $P(0)$ should be 3, and use timesteps 0 to 10 with a step of 0.1.

```python
import scipy.integrate
def degradation_model(t,y):
    return -2*y

sol = scipy.integrate.solve_ivp(fun=degradation_model, t_span=(0,10), y0=(3,), 
                                t_eval=np.arange(0,10,0.1) )

print( sol.t )
print( sol.t.shape )
print( sol.y.shape )
```

For the purposes of parameter-fitting, we want to avoid hard-coding the coefficients. I will refer to these as the model parameters or just "parameters". We should re-write the model to use parameters. And then supply them to solve_ivp in the "args" argument.

```python
def degradation_model2(t,y,params):
    return -params[0]*y

d1 = 2
params = (d1,)
sol = scipy.integrate.solve_ivp(fun=degradation_model2, t_span=(0,10), y0=(3,), 
                                t_eval=np.arange(0,10,0.1), args=(params,) )

print( sol.t )
print( sol.t.shape )
print( sol.y.shape )
```

## For older versions of Scipy

If you don't have solve_ivp, then we need to make a couple of adjustments. First swap the order of t and y in the model function. Then, update the call to odeint. the function is called "func" instead of "fun". It takes the entire t array in just one parameter. Also, it returns just the y values.

```python
def degradation_model2_old(y,t,params):
    return -params[0]*y

d1 = 2
params = (d1,)
t = np.arange(0,10,0.1)
y = scipy.integrate.odeint(func=degradation_model2_old, t=t, y0=(3,),args=(params,) )

print( t )
print( y )
print( y.shape )
```

```python
import matplotlib.pyplot as plt

plt.plot( t, y ) # notice that we don't need to transpose it.
plt.title( "simulation" )
plt.xlabel( 'Time')
plt.ylabel( "Concentration")
plt.legend( ["P"] );
```
